```python
import datetime as dt
import random
import pandas as pd
import numpy as np
import util as ut
import matplotlib.pyplot as plt
%matplotlib inline
import ManualStrategy as ms
import StrategyLearner as sl
from marketsimcode import compute_portvals
```

```python
def test_code():
    
    sd = dt.datetime(2008,1,1)
    ed = dt.datetime(2009,12,31)
    symbol = "JPM"
    moving_window = 21
    
    
    #Strategy Learner
   
    #Imapct 0.0
    learner = sl.StrategyLearner(verbose = False, impact = 0.0)
    learner.add_evidence(symbol="JPM",sd=sd,ed=ed,sv=100000)
    df_trades = learner.testPolicy(symbol="JPM",sd=sd,ed=ed,sv=100000)
    port_vals = compute_portvals(df_trades,start_val = 100000, commission = 0.0, impact = 0.0)
    port_vals_norm = port_vals/port_vals.iloc[0]
    
    #Imapct 0.002
    learner1 = sl.StrategyLearner(verbose = False, impact = 0.002)
    learner1.add_evidence(symbol="JPM",sd=sd,ed=ed,sv=100000)
    df_trades1 = learner1.testPolicy(symbol="JPM",sd=sd,ed=ed,sv=100000)
    port_vals1 = compute_portvals(df_trades1,start_val = 100000, commission = 0.0, impact = 0.002)
    port_vals_norm1 = port_vals1/port_vals1.iloc[0]
    
    #Imapct 0.005
    learner2 = sl.StrategyLearner(verbose = False, impact = 0.05)
    learner2.add_evidence(symbol="JPM",sd=sd,ed=ed,sv=100000)
    df_trades2 = learner2.testPolicy(symbol="JPM",sd=sd,ed=ed,sv=100000)
    port_vals2 = compute_portvals(df_trades2,start_val = 100000, commission = 0.0, impact = 0.05)
    port_vals_norm2 = port_vals2/port_vals2.iloc[0]
    
     
    
    fig,ax = plt.subplots(figsize=(12,7))
    ax = port_vals_norm.plot(color ='red', label = 'Impact 0.0')
    port_vals_norm1.plot(ax=ax,color ='green', label = 'Impact 0.002')
    port_vals_norm2.plot(ax=ax,color ='blue', label = 'Impact 0.05')


    plt.legend()
    plt.title("Strategy Learner")
#     plt.text(plt.xlim()[0] + 4, 
#     plt.ylim()[0] + 0.5 ,'AARTI LADDHA',  alpha=0.5,fontsize=40, color='gray',va='center')

    plt.show()
```

```python
if __name__ == "__main__":
    test_code()
```
