```python
import pandas as pd
import numpy as np
import os
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
import util as ut
from marketsimcode import compute_portvals
from indicators import getsma, getBBP, getMomentum
import random
import RTLearner as rt
import BagLearner as bl
from Manual_Strategy import *
```

```python
class StrategyLearner(object):  
    
    def __init__(self, verbose = False, impact =0.0, commission = 0.0):
        
        self.verbose = verbose
        self.impact = impact
        self.commission = commission
        self.learner = bl.BagLearner(learner = rt.RTLearner,kwargs={"leaf_size":5}, bags = 100, boost = False, verbose = False) 
        
    def add_evidence(self, symbol ="IBM", sd=dt.datetime(2008, 1, 1), ed=dt.datetime(2009, 1, 1),  sv=10000):
        
        moving_window = 21

        # example usage of the old backward compatible util function 
        syms = [symbol] 
        dates = pd.date_range(sd, ed)
        prices_all = ut.get_data(syms, dates)
        prices = prices_all[syms]
        prices_SPY = prices_all["SPY"]  # only SPY, for comparison later 

        #Getting Indicators values 
        sma, price_sma = getsma(prices, moving_window)
        upperband, lowerband, BBP = getBBP(prices, moving_window)
        momentum = getMomentum(prices, moving_window)

        sma_df = sma.rename(columns={symbol:'SMA'})
        bbp_df = BBP.rename(columns={symbol:'BBP'})
        momentum_df = momentum.rename(columns={symbol:'Momentum'})
        indicators_df = pd.concat((sma_df,bbp_df,momentum_df),axis=1)
        indicators_df.fillna(0, inplace = True)

        train_x = indicators_df[:-10]
        train_x = train_x.values

        # Constructing trainY
        train_y = []
        for i in range(prices.shape[0]-10):
           
            ratio = (prices.loc[prices.index[i+ 10],symbol]- prices.loc[prices.index[i],symbol])/prices.loc[prices.index[i],symbol]
            if ratio > (0.02 + self.impact):
                train_y.append(1)
            elif ratio < (-0.02 - self.impact):
                train_y.append(-1)
            else:
                train_y.append(0)
        train_y = np.array(train_y)

        # Training
        self.learner.add_evidence(train_x,train_y)
        
    def testPolicy(self, symbol ='IBM', sd=dt.datetime(2009,1,1), ed=dt.datetime(2010,1,1), sv = 10000):

        moving_window = 21
        syms = [symbol] 
        dates = pd.date_range(sd, ed)
        prices_all = ut.get_data(syms, dates)
        prices = prices_all[syms]

        sma, price_sma = getsma(prices, moving_window)
        upperband, lowerband, BBP = getBBP(prices, moving_window)
        momentum = getMomentum(prices, moving_window)

        sma_df=sma.rename(columns={symbol:'SMA'})
        bbp_df=BBP.rename(columns={symbol:'BBP'})
        momentum_df=momentum.rename(columns={symbol:'Momentum'})
        indicators_df = pd.concat((sma_df,bbp_df,momentum_df),axis=1)
        indicators_df.fillna(0, inplace = True)

        test_x =indicators_df.values

        # Querying the learner for testY
        test_y = self.learner.query(test_x)
        #print(test_y)

        # Constructing trades DataFrame
        trades_df = prices.copy()
        trades_df.loc[:]=0
        signal = 0

        for i in range(0, prices.shape[0]-1):
            index = prices.index[i]
            if signal == 0:
                if test_y[0][i] > 0 :
                    trades_df.loc[index, :] = 1000  
                    signal = 1

                elif test_y[0][i] < 0 :
                    trades_df.loc[index, :] = -1000 
                    signal = -1

            elif signal == -1:
                if test_y[0][i] > 0 :
                    trades_df.loc[index,:] = 2000
                    signal = 1

            elif signal == 1:
                if test_y[0][i] < 0: 
                    trades_df.loc[index,:] = - 2000
                    signal = -1

        if (signal == 1):
            trades_df.loc[index,:] = 1000
        else:
            trades_df.loc[index,:] = -1000
         
        return trades_df
```

```python
if __name__ == "__main__":  
    
    symbol =["JPM"]
    sd=dt.datetime(2008,1,1)
    ed=dt.datetime(2009,12,31)
    learner = StrategyLearner(verbose = False, impact=0.0)
    
    # Strategy Learner

    learner.add_evidence(symbol="JPM",sd=sd,ed=ed,sv=100000)
    trades_df_sl = learner.testPolicy(symbol="JPM",sd=sd,ed=ed,sv=100000)

    port_vals_sl = compute_portvals(trades_df_sl,start_val = 100000, commission = 9.95, impact = 0.005)
    
    #Normalizing PortFolio Values
    port_vals_norm_sl = port_vals_sl/port_vals_sl.iloc[0]

    cum_ret_man,avg_daily_ret_man,std_daily_ret_man,sharpe_ratio_man = Portfolio_Statistics(port_vals_norm_sl)
    
    print(f"Date Range: {sd} to {ed} for {symbol}")
    print()
    print("Manual Strategy")
    print(f"Sharpe Ratio of Fund: {sharpe_ratio_man}")  
    print(f"Cumulative Return of Fund: {cum_ret_man}") 
    print(f"Standard Deviation of Fund: {std_daily_ret_man}")
    print(f"Average Daily Return of Fund: {avg_daily_ret_man}") 
    print(f"Final Portfolio Value: {port_vals_sl[-1]}")
    
    fig,ax = plt.subplots(figsize=(15,7.5))
    ax = port_vals_norm_sl.plot(color ='red', label = 'Manual Strategy')
```
